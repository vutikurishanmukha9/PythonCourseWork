```python
import matplotlib.pyplot as plt

runs = [505, 973, 308, 530, 464, 466, 405, 341, 639, 741, 657]
seasons = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025]

plt.plot(seasons, runs,
         color='red',
         marker='o',
         linestyle='-',
         linewidth=1,   # fixed (was '1')
         label='Runs Scored By Kohli Over last 6 IPL Seasons')  # fixed typo

plt.title('Runs Scored By Kohli Over last 6 IPL Seasons')
plt.xlabel('Seasons')
plt.ylabel('Runs')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

ad_spend = [2, 4, 5, 7, 8, 9, 10, 11]

sales = [20, 25, 28, 35, 40, 48, 50, 52]

sizes = [200, 250, 280, 350, 400, 480, 500, 520]

colors = sales

plt.scatter(ad_spend, sales,
            s = sizes,
            c = colors,
            alpha = 0.8,
            edgecolors = 'black',
            linewidths = 1.2)

plt.title('Ad Spend vs Sales')
plt.xlabel('Ad Spend')
plt.ylabel('Sales')

plt.colorbar(label='sales values')

plt.grid(True, linestyle = "-", alpha = 0.2)
plt.show()
```

```python
import matplotlib.pyplot as plt

runs = [505, 973, 308, 530, 464, 466, 405, 341, 639, 741, 657]
seasons = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025]

plt.pie(
    runs,
    labels=seasons,
    autopct="%.1f%%",
    startangle=180,
    colors=["violet", "indigo", "blue", "green", "yellow", "orange", "red",
            "gold", "lightgreen", "cyan", "brown"],  
    explode=[0, 0, 0, 0, 0.05, 0.1, 0.1, 0.05, 0.05, 0.07, 0.1],
    shadow=True
)

plt.title("Kohli Runs")
plt.show()
```
