```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Super_Market.csv", encoding="latin1")

df["Order Date"] = pd.to_datetime(df["Order Date"], format='mixed')

sales_trend = df.groupby(df["Order Date"].dt.to_period("M"))["Sales"].sum()

sales_trend.plot(color='black', marker='s',linestyle='-',linewidth=1, alpha = 0.7)
plt.title("Monthly Sales Trend")
plt.ylabel("Total Sales")
plt.xlabel("Month")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("Super_Market.csv", encoding="latin1")
df["Order Date"] = pd.to_datetime(df["Order Date"], format='mixed')
df["Month-Year"] = df["Order Date"].dt.to_period("M")


monthly_profit = df.groupby("Month-Year")["Profit"].sum()

monthly_profit.plot(color='black', marker='s',linestyle='-',linewidth=1, alpha = 0.7)
plt.title("Monthly Profit Trends")
plt.xlabel("Month-Year")
plt.ylabel("Total Profit")
plt.show()
```

```python
import seaborn as sb
import matplotlib.pyplot as plt

category_sales = df.groupby("Category")["Sales"].sum().reset_index()

sb.barplot(data=category_sales, x="Category", y="Sales", color="black", edgecolor = 'red',alpha = 0.7)
plt.title("Total Sales per Product Category")
plt.ylabel("Sales")
plt.xlabel("Category")
plt.show()
```

```python
import matplotlib.pyplot as plt

top_customers = df.groupby("Customer Name")["Sales"].sum().nlargest(10)

plt.bar(top_customers.index, top_customers.values, color='black', edgecolor = 'red',alpha = 0.7)  
plt.title("Top 10 Customers by Sales")
plt.xlabel("Customer")
plt.ylabel("Total Sales")
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()
plt.show()
```

```python
import seaborn as sb
import matplotlib.pyplot as plt


region_sales = df.groupby("Region")["Sales"].sum().sort_values(ascending=False)

sb.barplot(x=region_sales.index, y=region_sales.values, color = "black", edgecolor = 'red', alpha = 0.7)
plt.title("Sales by Region")
plt.ylabel("Total Sales")
plt.show()
```

```python
import seaborn as sb
import matplotlib.pyplot as plt


shipmode_profit = df.groupby("Ship Mode")["Profit"].sum().sort_values(ascending=False)


sb.barplot(x=shipmode_profit.index, y=shipmode_profit.values, color = "black", edgecolor = 'red', alpha = 0.7)
plt.title("Profit by Ship Mode")
plt.ylabel("Total Profit")
plt.show()
```

```python
import matplotlib.pyplot as plt

category_sales = df.groupby("Category")["Sales"].sum()

plt.pie(
    category_sales,
    labels=category_sales.index,
    autopct="%.1f%%",
    startangle=180,
    colors=plt.cm.Set3.colors,   
    explode=[0.05]*len(category_sales),  
    shadow=True
)

plt.title("Sales Distribution by Category")
plt.show()
```

```python
import matplotlib.pyplot as plt

import seaborn as sb

sb.scatterplot(
    data=df,
    x="Discount",
    y="Profit",
    alpha=0.6, 
    color="black"  
)


plt.title("Discount vs Profitability", fontsize=16)
plt.xlabel("Discount (%)", fontsize=12)
plt.ylabel("Profit ($)", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.hist(df["Sales"], bins=50, color='black', edgecolor='red', alpha = 0.7)  
plt.title('Sales Frequency Distribution')
plt.xlabel('Sales Amount')
plt.ylabel('Frequency')
plt.show()
```
