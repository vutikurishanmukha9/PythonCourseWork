```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Super_Market.csv", encoding="latin1")

df["Order Date"] = pd.to_datetime(df["Order Date"], format='mixed')

sales_trend = df.groupby(df["Order Date"].dt.to_period("M"))["Sales"].sum()

sales_trend.plot(kind="line", marker="s", color="black")
plt.title("Monthly Sales Trend")
plt.ylabel("Total Sales")
plt.xlabel("Month")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("Super_Market.csv", encoding="latin1")
df["Order Date"] = pd.to_datetime(df["Order Date"], format='mixed')
df["Month-Year"] = df["Order Date"].dt.to_period("M")


monthly_profit = df.groupby("Month-Year")["Profit"].sum()

monthly_profit.plot(kind="line", marker="s", color="black")
plt.title("Monthly Profit Trends")
plt.xlabel("Month-Year")
plt.ylabel("Total Profit")
plt.show()
```

```python
import seaborn as sb
import matplotlib.pyplot as plt

category_sales = df.groupby("Category")["Sales"].sum().reset_index()

sb.barplot(data=category_sales, x="Category", y="Sales", palette="Set1")
plt.title("Total Sales per Product Category")
plt.ylabel("Sales")
plt.xlabel("Category")
plt.show()
```

```python
import seaborn as sb
import matplotlib.pyplot as plt

top_customers = df.groupby("Customer Name")["Sales"].sum().nlargest(10)

sb.barplot(x=top_customers.values, y=top_customers.index, palette="plasma")
plt.title("Top 10 Customers by Sales")
plt.xlabel("Total Sales")
plt.ylabel("Customer")
plt.show()
```

```python
import seaborn as sb
import matplotlib.pyplot as plt


region_sales = df.groupby("Region")["Sales"].sum().sort_values(ascending=False)

sb.barplot(x=region_sales.index, y=region_sales.values, palette="plasma")
plt.title("Sales by Region")
plt.ylabel("Total Sales")
plt.show()
```

```python
import seaborn as sb
import matplotlib.pyplot as plt


shipmode_profit = df.groupby("Ship Mode")["Profit"].sum().sort_values(ascending=False)


sb.barplot(x=shipmode_profit.index, y=shipmode_profit.values, palette="plasma")
plt.title("Profit by Ship Mode")
plt.ylabel("Total Profit")
plt.show()
```
